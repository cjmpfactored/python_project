```python
# pearson correlation matrix
f = plt.figure(figsize=(19, 15))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14, rotation=45)
plt.yticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```

```python
# calculating two correlation tables
df_corr1 = df.corr(method ='pearson')
df_corr2 = df.corr(method ='spearman')
```

```python
# joining both correlations for response variable together
correlations = pd.concat([df_corr1['RET_M1'], df_corr2['RET_M1']], axis = 1)
correlations.columns = ['PEARSON_CORRELATION_RET_M1', 'SPEARMAN_CORRELATION_RET_M1']
correlations
```

```python
# filtering variables with more than .1 or less than -.1 correlations (both correlations)

correlations_mask = ((correlations['PEARSON_CORRELATION_RET_M1'] > 0.1) | (correlations['PEARSON_CORRELATION_RET_M1'] < -0.1)) & ((correlations['SPEARMAN_CORRELATION_RET_M1'] > 0.1) | (correlations['SPEARMAN_CORRELATION_RET_M1'] < -0.1))
chosen_correlations = correlations[correlations_mask]
chosen_correlations
```

```python
chosen_variables_1 = chosen_correlations.index.values.tolist()  # an ndarray method, you probably shouldn't depend on this
chosen_variables_1 
```
